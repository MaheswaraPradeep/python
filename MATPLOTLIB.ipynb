```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([88,86,100,97,86,95])
a=["English","VDC","MECH","PHYSICS","C LANGUAGE","MATHS"]
plt.pie(y,labels=a)
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([88,86,100,97,86,95])
a=["English","VDC","MECH","PHYSICS","C LANGUAGE","MATHS"]
plt.pie(y,labels=a,startangle=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([88,86,100,97,86,95])
a=["English","VDC","MECH","PHYSICS","C LANGUAGE","MATHS"]
exp=[0,0,0.3,0.1,0,0.1]
plt.pie(y,labels=a,explode=exp,shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([88,86,100,97,86,95])
a=["English","VDC","MECH","PHYSICS","C LANGUAGE","MATHS"]
col=["m","green","blue","hotpink","y","c"]
plt.pie(y,labels=a,colors=col)
plt.show()
```
